```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import warnings
from sklearn.exceptions import ConvergenceWarning

def load_face_database(path, target_size=(100, 100)):
    images = []
    labels = []
    label_names = sorted(os.listdir(path))
    print(f"Found labels: {label_names}")
    for label in label_names:
        if label.startswith('.'):  # skip system files like .DS_Store
            continue
        label_path = os.path.join(path, label)
        if not os.path.isdir(label_path):
            continue
        for filename in os.listdir(label_path):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.pgm')):
                img_path = os.path.join(label_path, filename)
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                if img is not None:
                    img_resized = cv2.resize(img, target_size)
                    images.append(img_resized)
                    labels.append(label)
    images = np.array(images)
    labels = np.array(labels)
    print(f"Loaded {len(images)} images from {len(set(labels))} classes.")
    print(f"Image size: {target_size}, Total dataset size: {images.shape}")
    return images, labels

def display_sample_faces(images, labels, num_samples=10):
    plt.figure(figsize=(15, 5))
    for i in range(min(num_samples, len(images))):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(labels[i])
        plt.axis('off')
    plt.suptitle(f"Sample {num_samples} Face Images", fontsize=16)
    plt.show()

def display_eigenfaces(pca, h, w, num_components=10):
    eigenfaces = pca.components_[:num_components]
    plt.figure(figsize=(15, 5))
    for i in range(num_components):
        plt.subplot(1, num_components, i + 1)
        plt.imshow(eigenfaces[i].reshape((h, w)), cmap='gray')
        plt.title(f"Eigenface {i+1}")
        plt.axis('off')
    plt.suptitle(f"Top {num_components} Eigenfaces", fontsize=16)
    plt.show()

def evaluate_model(images, labels, num_components=150, test_size=0.4, random_state=42):
    h, w = images.shape[1], images.shape[2]
    X = np.array([img.flatten() for img in images])
    le = LabelEncoder()
    y = le.fit_transform(labels)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, stratify=y, random_state=random_state)

    print(f"\nTraining set size: {X_train.shape[0]} samples")
    print(f"Testing set size: {X_test.shape[0]} samples")

    # PCA
    pca = PCA(n_components=num_components, whiten=True, random_state=random_state)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    print(f"Extracted top {num_components} eigenfaces")

    # Fisherfaces using LDA
    lda = LDA()
    X_train_lda = lda.fit_transform(X_train_pca, y_train)
    X_test_lda = lda.transform(X_test_pca)
    print(f"Computed Fisherfaces with LDA")

    # MLP Classifier
    warnings.filterwarnings("ignore", category=ConvergenceWarning)
    mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, early_stopping=True, random_state=random_state)
    mlp.fit(X_train_lda, y_train)
    y_pred = mlp.predict(X_test_lda)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Trained MLP Classifier. Accuracy: {accuracy:.4f}")

    return pca, lda, mlp, X_test_lda, y_test, le, h, w

# ---- MAIN ----
dataset_path = 'dataset/att_faces'  

face_db, labels = load_face_database(dataset_path)
display_sample_faces(face_db, labels, num_samples=10)

pca_model, lda_model, mlp_model, X_test_lda, y_test, label_encoder, h, w = evaluate_model(
    face_db, labels, num_components=150)

display_eigenfaces(pca_model, h, w, num_components=10)
```
